```python
import matplotlib.pyplot as plt

# Data for Madhya Pradesh and Rajasthan elections
parties_mp = ['BJP', 'INC', 'BSP', 'Others']
seats_mp = [163, 66, 0, 1]
percentages_mp = [round(seat / sum(seats_mp) * 100, 2) for seat in seats_mp]

parties_rj = ['INC', 'BJP', 'BSP', 'Others']
seats_rj = [69, 115, 2, 13]
percentages_rj = [round(seat / sum(seats_rj) * 100, 2) for seat in seats_rj]

# Create a figure with two subplots for pie charts
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

# Pie chart for Madhya Pradesh
ax1.pie(seats_mp, labels=[f"{party}: {percent}%" for party, percent in zip(parties_mp, percentages_mp)],
        autopct='%1.1f%%', )
ax1.set_title('Madhya Pradesh Seat Share')

# Pie chart for Rajasthan
ax2.pie(seats_rj, labels=[f"{party}: {percent}%" for party, percent in zip(parties_rj, percentages_rj)],
        autopct='%1.1f%%', )
ax2.set_title('Rajasthan Seat Share')


# Bar chart for both states
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
index = range(len(parties_mp))

# Plot bar chart
ax.bar(index, seats_mp, bar_width, label='Madhya Pradesh', color='blue')
ax.bar([i + bar_width for i in index], seats_rj, bar_width, label='Rajasthan', color='orange')

ax.set_xlabel('Parties')
ax.set_ylabel('Seats Won')
ax.set_title('Assembly Elections 2023: Seat Distribution')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(parties_mp)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import random
import math
import matplotlib.pyplot as plt

# User input for K and N
K = max(10, int(input("Enter the number of random numbers (minimum 10): ")))
N = int(input("Enter the upper limit for random numbers: "))

# Generate K random numbers within the limit N
random_numbers = [random.randint(1, N) for _ in range(K)]

# Function to check if a number is prime
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return False
    return True

# Separate prime and composite numbers
def separate_numbers(numbers):
    primes = [num for num in numbers if is_prime(num)]
    composites = [num for num in numbers if not is_prime(num) and num > 1]
    return primes, composites

# Get primes and composites
primes, composites = separate_numbers(random_numbers)

# Calculate squares and square roots
prime_squares = [p ** 2 for p in primes]
composite_roots = [math.sqrt(c) for c in composites]

# Plot scatter plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

ax1.scatter(primes, prime_squares, color='green', label='Prime Numbers')
ax1.set_title('Prime Numbers vs Squares')
ax1.set_xlabel('Prime Numbers')
ax1.set_ylabel('Squares')
ax1.legend()

ax2.scatter(composites, composite_roots, color='red', label='Composite Numbers')
ax2.set_title('Composite Numbers vs Square Roots')
ax2.set_xlabel('Composite Numbers')
ax2.set_ylabel('Square Roots')
ax2.legend()

plt.tight_layout()
plt.show()
```
